```python
import pandas as pd

df = pd.read_csv('test.csv')

df.head()
```

```python
# id별로 행의 개수와 value의 합계를 구함
id_stats = df.groupby('AgitID').agg(
    row_count=('AgitID', 'size'),  # 행의 개수
)

print(id_stats)
```

```python
df_group = pd.read_csv('test_group.csv')
df_group
```

```python
merged_df = pd.merge(id_stats, df_group, on='AgitID', how='left')

# 결과 확인
print(merged_df)
```
